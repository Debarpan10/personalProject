# Machine Learning

```python
D=df.groupby('Year').Total.sum()
D
```

```python
x=D.index.values.reshape(-1,1)
x
```

```python
y=D.values
y
```

```python
plt.scatter(x,y)
plt.xlabel("Year")
plt.ylabel('Total Suicides')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(x,y)
```

```python
#Slope(m)
LR.coef_
```

```python
#Intercept(c)
LR.intercept_
```

```python
y_pred=LR.predict(x)
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred)
plt.show()
```

```python
LR.score(x,y) * 100
```

```python
LR.predict([[2020]])
```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=5)
X=poly.fit_transform(x)
pd.DataFrame(X)
```

```python
from sklearn.linear_model import LinearRegression
Regr=LinearRegression()
Regr.fit(X,y)
```

```python
Regr.coef_
```

```python
Regr.intercept_
```
