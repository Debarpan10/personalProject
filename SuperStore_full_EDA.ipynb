```python
# Function to remove outliers using the interquartile range (IQR)
def remove_outliers(df, column, threshold=1.5):
    q1 = df['Sales'].quantile(0.25)
    q3 = df['Sales'].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - threshold * iqr
    upper_bound = q3 + threshold * iqr
    df_no_outliers = df[(df['Sales'] >= lower_bound) & (df['Sales'] <= upper_bound)]
    return df_no_outliers

# Apply the function to remove outliers from column 'Sales'
df_no_outliers = remove_outliers(df, 'Sales')
# Display the DataFrame without outliers
print(df_no_outliers)
```

```python
# Function to remove outliers using the interquartile range (IQR)
def remove_outliers(df, column, threshold=1.5):
    q1 = df['Profit'].quantile(0.25)
    q3 = df['Profit'].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - threshold * iqr
    upper_bound = q3 + threshold * iqr
    df_no_outliers_p = df[(df['Profit'] >= lower_bound) & (df['Profit'] <= upper_bound)]
    return df_no_outliers_p

# Apply the function to remove outliers from column 'Profit'
df_no_outliers_p = remove_outliers(df, 'Profit')
# Display the DataFrame without outliers
print(df_no_outliers_p)
```
