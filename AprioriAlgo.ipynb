```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
df=pd.read_csv("D://Rm_databases//store.csv",header=None)
df
```

```python
df.shape
```

Step5:Convert Pandas Dataframe into a list of lists

```python
records=[]
for i in range(0,22):
    records.append([str(df.values[i,j]) for j in range(0,6)])
```

Step6: Build the Apriori Model

```python
#Building the first apriori model3
association_rules=apriori(records,min_support=0.50,min_confidence=0.7,min_lift=1.2,min_length=2)
association_results=list(association_rules)
```

Step7: Print out the number of rules

```python
print(association_results)
```
